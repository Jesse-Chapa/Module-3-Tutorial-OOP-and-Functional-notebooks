## Exercise 1 — Sort an Array of 0s, 1s, and 2s (OOP Style)

This exercise uses an object-oriented programming structure with a class method.  
The goal is to sort an array that contains only the values 0, 1, and 2 without using Python’s built-in sort functions.

I implemented the Dutch National Flag algorithm, which:
- Uses three pointers (low, mid, high)
- Makes only one pass through the list
- Uses constant extra memory
- Swaps values in place

This approach is efficient and meets the problem constraint requirements.


```python
class Solution:
    def sort012(self, arr):
        low = 0
        mid = 0
        high = len(arr) - 1

        while mid <= high:
            if arr[mid] == 0:
                arr[low], arr[mid] = arr[mid], arr[low]
                low += 1
                mid += 1
            elif arr[mid] == 1:
                mid += 1
            else:
                arr[mid], arr[high] = arr[high], arr[mid]
                high -= 1
```
