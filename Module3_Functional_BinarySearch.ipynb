## Exercise 2 â€” Binary Search (Functional Style)

This exercise demonstrates a functional algorithm approach using binary search on a sorted list.

Binary search works by:
- Repeatedly dividing the search interval in half
- Comparing the middle value to the target
- Adjusting the search boundaries accordingly

My implementation:
- Uses an iterative loop
- Returns the index if the value is found
- Returns -1 if not found
- Continues searching left to return the first (smallest) index when duplicates exist

This provides O(log n) time complexity and avoids unnecessary scanning.


```python
class Solution:
    def binarysearch(self, arr, k):
          left = 0
          right = len(arr) - 1
          result = -1

          while left <= right:
            mid = (left + right) // 2

            if arr[mid] == k:
                result = mid
                right = mid - 1
            elif arr[mid] < k:
                left = mid + 1
            else:
                right = mid - 1

          return result
        
```
